```
import numpy as np

# Ensure df_distribucion_quantities is available. If not, create a dummy for demonstration.
if 'df_distribucion_quantities' not in locals() or df_distribucion_quantities is None:
    print("df_distribucion_quantities not found, creating a dummy DataFrame for adjustment.")
    data = {
        'Referencia': ['REF1', 'REF2', 'REF3'],
        'Cantidades en sistema': [100.0, 150.5, 60.2],
        'Store A': [10.3, 15.1, 6.0],
        'Store B': [20.7, 30.9, 12.1],
        'Store C': [30.5, 45.2, 18.0],
        'Store D': [38.5, 59.3, 24.1]
    }
    df_distribucion_quantities = pd.DataFrame(data)

# Identify the store quantity columns
# Assuming all columns except 'Referencia' and 'Cantidades en sistema' are store columns
store_columns = [col for col in df_distribucion_quantities.columns if col not in ['Referencia', 'Cantidades en sistema']]

# Iterate through each row to adjust quantities
for index, row in df_distribucion_quantities.iterrows():
    original_total_quantity = row['Cantidades en sistema']

    # Calculate the sum of current floating-point quantities
    current_distributed_sum = row[store_columns].sum()

    # Calculate the difference
    initial_difference = original_total_quantity - current_distributed_sum

    # Round each quantity to the nearest integer
    rounded_quantities = row[store_columns].apply(lambda x: int(round(x)))

    # Calculate the sum of rounded quantities
    rounded_distributed_sum = rounded_quantities.sum()

    # Calculate the difference after rounding
    remaining_difference = int(original_total_quantity) - rounded_distributed_sum

    # Create a list of stores that had a non-zero quantity after rounding, prioritizing for redistribution
    stores_with_non_zero_rounded = rounded_quantities[rounded_quantities > 0].index.tolist()

    # Distribute the remaining difference (add or subtract 1)
    dist_index = 0
    while remaining_difference != 0 and stores_with_non_zero_rounded:
        store_name = stores_with_non_zero_rounded[dist_index % len(stores_with_non_zero_rounded)]

        if remaining_difference > 0:
            rounded_quantities[store_name] += 1
            remaining_difference -= 1
        else: # remaining_difference < 0
            if rounded_quantities[store_name] > 0: # Only subtract if quantity is > 0
                 rounded_quantities[store_name] -= 1
                 remaining_difference += 1
        dist_index += 1
        # If we've gone through all candidates and still have a difference, we need to adjust the strategy.
        # For this subtask, we'll stick to adjusting among those with non-zero rounded quantities.
        # If remaining_difference is still not 0 after iterating through all candidates,
        # it means the remaining difference cannot be distributed among stores with non-zero rounded quantities
        # without making some quantities negative or changing stores that should be zero.
        # We will break the loop in such cases to avoid infinite loops and potentially indicate an issue
        # in the initial distribution or constraints.

        # Simple check to prevent infinite loop if remaining_difference is not zero but no valid candidates left
        if remaining_difference != 0 and dist_index >= len(stores_with_non_zero_rounded) * 2: # Check after two full rounds
             print(f"Warning: Could not perfectly balance distribution for row {index}. Remaining difference: {remaining_difference}")
             break


    # Ensure no adjusted quantity is negative
    for store_name in store_columns:
        if rounded_quantities[store_name] < 0:
            rounded_quantities[store_name] = 0

    # Update the store quantity columns in the original DataFrame
    df_distribucion_quantities.loc[index, store_columns] = rounded_quantities.values


# Convert store quantity columns to integer type after adjustment
for store_name in store_columns:
    df_distribucion_quantities[store_name] = df_distribucion_quantities[store_name].astype(int)

# Verify the sum of adjusted quantities for each row (optional, but good practice)
df_distribucion_quantities['adjusted_sum_verification'] = df_distribucion_quantities[store_columns].sum(axis=1)
print("\nDataFrame after adjustment:")
display(df_distribucion_quantities.head())

# Optionally, check if the adjusted sum equals the original total quantity
# Note: Due to rounding and the distribution method, the sum might not exactly match the original float total,
# but it should match the rounded integer total if the logic worked correctly.
# Let's compare against the rounded integer total
df_distribucion_quantities['original_total_rounded'] = df_distribucion_quantities['Cantidades en sistema'].apply(lambda x: int(round(x)))
verification_successful = (df_distribucion_quantities['adjusted_sum_verification'] == df_distribucion_quantities['original_total_rounded']).all()
print(f"\nVerification successful (adjusted sum == rounded original total): {verification_successful}")

# Drop verification columns
df_distribucion_quantities = df_distribucion_quantities.drop(columns=['adjusted_sum_verification', 'original_total_rounded'])
```

I need to implement the distribution logic for each of the first 5 quantities, applying minimums, rounding, and redistributing the remainder while prioritizing Grande and Mediano stores and ensuring the final quantities are integers that sum up to the original total.



```
import numpy as np

# Ensure df_stores and store_counts_by_format are available from previous steps
if 'df_stores' not in locals() or 'store_counts_by_format' not in locals():
    print("Required DataFrames or variables are not available. Please run previous steps.")
else:
    minimum_quantity = 6
    final_distribution_results = {}

    # Ensure first_5_quantities is available
    if 'first_5_quantities' not in locals() or not first_5_quantities:
        print("First 5 quantities not found. Please run the step to select them.")
    else:
        # Get separate lists of prioritized and non-prioritized store names
        prioritized_store_names = df_stores[df_stores['Formato'].isin(['Grande', 'Mediano'])]['Nombre'].tolist()
        non_prioritized_store_names = df_stores[df_stores['Formato'] == 'Pequeño']['Nombre'].tolist()
        all_store_names = prioritized_store_names + non_prioritized_store_names


        # Loop through each of the first 5 quantities
        for quantity in first_5_quantities:
            adjusted_quantities_for_this_quantity = {}
            total_remainder = 0.0 # Remainder from not meeting minimums (based on initial floats)

            # Calculate initial quantity per store by format
            initial_quantity_by_format = {}
            for store_format, count in store_counts_by_format.items():
                if count > 0:
                    initial_quantity_by_format[store_format] = float(quantity) / count
                else:
                    initial_quantity_by_format[store_format] = 0.0

            # Apply minimum quantity rule and round to integers initially
            initial_distributed_sum = 0
            stores_that_met_minimum = [] # Stores that initially received >= minimum after rounding

            for index, store in df_stores.iterrows():
                store_name = store['Nombre']
                store_format = store['Formato']

                # Get the initial calculated quantity for this store's format
                initial_quantity = initial_quantity_by_format.get(store_format, 0.0)

                # Round the initial quantity to the nearest integer
                rounded_quantity = int(round(initial_quantity))

                # Apply the minimum quantity rule
                if rounded_quantity < minimum_quantity:
                    adjusted_quantities_for_this_quantity[store_name] = 0
                    # Add the initial quantity (before rounding) to the remainder if it was less than minimum
                    total_remainder += initial_quantity
                else:
                    adjusted_quantities_for_this_quantity[store_name] = rounded_quantity
                    initial_distributed_sum += rounded_quantity
                    stores_that_met_minimum.append(store_name)


            # Calculate remaining integer quantity to distribute to match the original total
            # We will distribute the difference between the rounded original total and the sum of initially assigned integers
            remaining_integer_difference = int(round(quantity)) - initial_distributed_sum

            # Distribute the remaining integer difference in a round-robin fashion
            # among stores that met the minimum, prioritizing Grande and Mediano.
            distribution_order_balancing = [name for name in prioritized_store_names if name in stores_that_met_minimum] + \
                                           [name for name in non_prioritized_store_names if name in stores_that_met_minimum]

            dist_index = 0
            while remaining_integer_difference != 0 and distribution_order_balancing:
                store_name = distribution_order_balancing[dist_index % len(distribution_order_balancing)]

                if remaining_integer_difference > 0:
                    adjusted_quantities_for_this_quantity[store_name] += 1
                    remaining_integer_difference -= 1
                else: # remaining_integer_difference < 0
                    # Only subtract if the current quantity for the store is greater than the minimum (6)
                    # to avoid going below the minimum after balancing
                    if adjusted_quantities_for_this_quantity[store_name] > minimum_quantity:
                        adjusted_quantities_for_this_quantity[store_name] -= 1
                        remaining_integer_difference += 1
                    # If quantity is already at minimum, skip this store for subtraction in this round


                dist_index += 1


            # Ensure all adjusted quantities are non-negative integers
            for store_name in adjusted_quantities_for_this_quantity.keys():
                if adjusted_quantities_for_this_quantity[store_name] < 0:
                    adjusted_quantities_for_this_quantity[store_name] = 0
                adjusted_quantities_for_this_quantity[store_name] = int(adjusted_quantities_for_this_quantity[store_name])


            # Store the adjusted quantities for this quantity
            final_distribution_results[quantity] = adjusted_quantities_for_this_quantity

        # Print the final distribution results for each quantity
        print("\nFinal Distribution Results for the first 5 quantities:")
        for quantity, distribution in final_distribution_results.items():
            print(f"\nOriginal Quantity: {quantity}")
            # Print the distribution for each store
            for store_name, adjusted_qty in distribution.items():
                 print(f"  {store_name}: {adjusted_qty}")

            # Optional: Verify the sum for each quantity
            final_sum_for_quantity = sum(distribution.values())
            print(f"  Sum of distributed quantities: {final_sum_for_quantity}")
            print(f"  Original quantity (rounded integer): {int(round(quantity))}")
            print(f"  Difference (should be 0): {int(round(quantity)) - final_sum_for_quantity}")
```

```
import numpy as np
import pandas as pd

# Define minimum quantities (general minimum for this task is 6)
minimum_quantity = 6

# Ensure df_stores and store_counts_by_format are available from previous steps
if 'df_stores' not in locals() or 'store_counts_by_format' not in locals():
    print("Required DataFrames or variables are not available. Please run previous steps.")
else:
    # Ensure all_quantities is available
    if 'all_quantities' not in locals() or not all_quantities:
        print("All quantities not found or empty. Please run the step to select them.")
    else:
        # Create a DataFrame from all quantities for easier vectorized operations
        # Process ALL quantities now as requested
        df_quantities_to_distribute = pd.DataFrame(all_quantities, columns=['Original Quantity'])


        # Get separate lists of prioritized and non-prioritized store names
        prioritized_store_names = df_stores[df_stores['Formato'].isin(['Grande', 'Mediano'])]['Nombre'].tolist()
        non_prioritized_store_names = df_stores[df_stores['Formato'] == 'Pequeño']['Nombre'].tolist()
        all_store_names = prioritized_store_names + non_prioritized_store_names

        # Prepare DataFrame to store adjusted quantities for all stores
        df_adjusted_distribution = df_quantities_to_distribute.copy()
        for store_name in all_store_names:
            df_adjusted_distribution[store_name] = 0.0 # Initialize with float for calculations


        # --- Distribution Logic with Attempt for Equal Quantities within Format (Vectorized where possible) ---

        # Calculate initial theoretical quantity for each store based on original participation
        for store_name in all_store_names:
             store_participation = df_stores.loc[df_stores['Nombre'] == store_name, 'Participacion'].iloc[0]
             df_adjusted_distribution[store_name] = df_adjusted_distribution['Original Quantity'] * store_participation


        # Apply minimum quantity rule (6 for all stores in this task) and round to integers initially (vectorized)
        initial_distributed_sum_series = pd.Series(0.0, index=df_adjusted_distribution.index)

        for store_name in all_store_names:
            min_qty = minimum_quantity # Minimum is 6 for all stores in this specific task

            # Round the initial quantity to the nearest integer
            rounded_quantity_series = np.round(df_adjusted_distribution[store_name]).astype(int)

            # Apply the minimum quantity rule: if rounded < min, set to 0
            df_adjusted_distribution[store_name] = np.where(rounded_quantity_series < min_qty, 0, rounded_quantity_series)

            # Sum up the initially assigned quantities (those >= minimum)
            initial_distributed_sum_series += np.where(rounded_quantity_series >= min_qty, rounded_quantity_series, 0)


        # Calculate remaining integer quantity to distribute to match the original total (vectorized)
        df_adjusted_distribution['Original Quantity'] = df_adjusted_distribution['Original Quantity'].astype(int)
        remaining_integer_difference_series = df_adjusted_distribution['Original Quantity'] - initial_distributed_sum_series


        # Distribute the remaining integer difference in a round-robin fashion (row-wise loop for final balancing)
        # This part is still row-wise to ensure accurate integer distribution and total matching.
        # The optimization comes from vectorizing the initial calculation and minimum application.

        final_distribution_results = {} # Dictionary to store results for creating the final table

        for index, row in df_adjusted_distribution.iterrows():
             quantity = row['Original Quantity']
             adjusted_quantities_for_this_quantity = {store_name: row[store_name] for store_name in all_store_names}

             current_distributed_sum = sum(adjusted_quantities_for_this_quantity.values())
             remaining_difference = quantity - current_distributed_sum

             # Distribute the remaining difference in a round-robin fashion
             stores_with_non_zero_adjusted = [name for name, qty in adjusted_quantities_for_this_quantity.items() if qty > 0]

             # Prioritize Grande and Mediano stores for distributing remaining difference
             # Also prioritize stores with higher initial quantity within prioritized/non-prioritized groups
             distribution_order_balancing = sorted(stores_with_non_zero_adjusted,
                                                    key=lambda store_name: (store_name not in prioritized_store_names,
                                                                            adjusted_quantities_for_this_quantity[store_name]), # Prioritize higher quantities among non-zero adjusted
                                                    reverse=False)


             dist_index = 0
             while remaining_difference != 0 and distribution_order_balancing:
                 store_name = distribution_order_balancing[dist_index % len(distribution_order_balancing)]

                 if remaining_difference > 0:
                     adjusted_quantities_for_this_quantity[store_name] += 1
                     remaining_difference -= 1
                 else: # remaining_difference < 0
                     # Only subtract if the current quantity for the store is greater than the minimum (6)
                     if adjusted_quantities_for_this_quantity[store_name] > minimum_quantity:
                         adjusted_quantities_for_this_quantity[store_name] -= 1
                         remaining_difference += 1
                     # If quantity is already at minimum, skip this store for subtraction in this round


                 dist_index += 1

             # Ensure all adjusted quantities are non-negative integers
             for store_name in all_store_names:
                 if adjusted_quantities_for_this_quantity[store_name] < 0:
                     adjusted_quantities_for_this_quantity[store_name] = 0
                 # Quantities should already be integers from rounding and adding/subtracting 1


             # Store the adjusted quantities for this quantity
             final_distribution_results[quantity] = adjusted_quantities_for_this_quantity


        # Create the final DataFrame from final_distribution_results dictionary
        original_quantities_list = sorted(final_distribution_results.keys())

        data_rows = []
        for quantity in original_quantities_list:
            adjusted_quantities_for_this_quantity = final_distribution_results[quantity]
            row_data = [quantity]

            for store_name in all_store_names:
                row_data.append(adjusted_quantities_for_this_quantity.get(store_name, 0))

            data_rows.append(row_data)

        store_columns_with_format = [f"{row['Nombre']} ({row['Formato']})" for index, row in df_stores.iterrows()]
        final_df_columns = ['Original Quantity'] + store_columns_with_format
        df_tabulated_distribution = pd.DataFrame(data_rows, columns=final_df_columns)


        # Print the head of the final tabulated DataFrame
        print("\nTabulated Distribution Table (Vectorized Approach - All Quantities):")
        display(df_tabulated_distribution.head())

        # Print the full table as well
        print("\nFull Tabulated Distribution Table (Vectorized Approach - All Quantities):")
        display(df_tabulated_distribution)
```
